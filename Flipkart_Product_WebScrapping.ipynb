```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.flipkart.com/search?q=rice+cooker&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY'
```

```python
web_page = requests.get(url)
web_page
```

```python
html_code = BeautifulSoup(web_page.content,'html.parser')
html_code
```

```python
cooker_list = []
```

```python
for items in html_code.find_all('div', class_='slAVV4'):
   prod_name = items.find('a', class_='wjcEIp').get('title')
   capacity = items.find('div',class_='NqpwHC').text
   prod_rating_element =items.find('div',class_="XQDdHH")
   if prod_rating_element:
      prod_rating = prod_rating_element.text
   else:
       prod_rating = '0.0'
   rating_count_element = items.find('span',class_='Wphh3N')
   if rating_count_element:
      rating_count = rating_count_element.text
   else:
       rating_count = '0.0'
   actual_price_element = items.find('div',class_='yRaY8j')
   if actual_price_element:
      #elements = actual_price_element.text.split('"')
       actual_price = actual_price_element.text
   else:
       actual_price = '0.0'
   current_price = items.find('div',class_='Nx9bqj').text
   current_price = current_price[1:]
   discount_element = items.find('div',class_='UkUFwK')
   if discount_element:
      #elements = actual_price_element.text.split('"')
       discount = discount_element.text
   else:
       discount = '0%'
   
   cookerdet_info = {
                'Product Name': prod_name,
                'Product Capacity': capacity,
                'Product Rating': prod_rating,
                'No Of Users Rated': rating_count,
                'Actual Price':actual_price[1:],
                'Current Price': current_price,
                'Discount': discount[:3]}
   #cooker_list.append([prod_name,capacity,prod_rating,rating_count,prod_price])
   cooker_list.append(cookerdet_info)
cooker_list
```

```python
cooker_df = pd.DataFrame(cooker_list)
cooker_df['No Of Users Rated'] = cooker_df['No Of Users Rated'].str.replace(r'[()]', '', regex=True)
cooker_df.head()
```

```python
cooker_df.to_csv(r'd:\RickCooker_Details_Flipkart.csv')
```
